```python
x = np.random.normal(size=100)
x
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x.cumsum());
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
y = np.random.normal(size=100)
```

```python
plt.plot(y.cumsum())
```

```python
plt.plot(x.cumsum(), label='x');
plt.plot(y.cumsum(), color='purple', label='y');
plt.legend(loc='upper left');
```

```python
# figure holds one or more axes.
```

```python
fig, axes = plt.subplots();
axes.plot(x.cumsum(), label='x');
axes.plot(y.cumsum(), color='purple', label='y');
axes.legend();

fig.set_size_inches(10,2)
```

### Bar Chart

```python
marathon = pd.read_csv('https://andybek.com/pandas-marathon')
marathon.head()
```

```python
# marathon runners by age
marathon.Age.plot();
```

```python
ax = marathon.Age.value_counts(sort=False).plot(kind='bar', figsize=(15,9))
ax.set_xlabel('Age Group');
ax.set_ylabel('# of runners');
```

```python
ax = marathon.Age.value_counts(ascending=True).plot(kind='barh', figsize=(15,9))
ax.set_xlabel('# of runners');
ax.set_ylabel('Age Group');
```

```python
# marathon runners by gender? 
```

```python
marathon['M/F'].value_counts(normalize=True).plot(kind='bar')
```

### Pie Chart

```python
games = pd.read_csv('https://andybek.com/pandas-games')
```

```python
games.head()
```

```python
# What is the market share for each platform in NA? 
games.groupby('Platform').NA_Sales.sum(). \
plot(kind='pie', subplots=True, figsize=(12,6))
```

```python
# does PS have a larger market share in JP ? 
games.groupby('Platform').JP_Sales.sum().plot(kind='pie')
```

```python
games.loc[:, ['Platform', 'NA_Sales', 'JP_Sales']] \
    .groupby('Platform').sum() \
    .plot(kind='pie', y='NA_Sales')
```

```python
games.loc[:, ['Platform', 'NA_Sales', 'JP_Sales']] \
    .groupby('Platform').sum() \
    .plot(kind='pie', subplots=True, figsize=(15,9))
```

### Scatter Plots

```python
sat = pd.read_csv('https://andybek.com/pandas-sat')
sat.head()
```

```python
# is reading and writing more strongly associated than writing and math? 
```

```python
math = sat[sat['SAT Section']=='Math'].Score.reset_index(drop=True)
math
```

```python
reading = sat[sat['SAT Section']=='Reading'].Score.reset_index(drop=True)
writing = sat[sat['SAT Section']=='Writing'].Score.reset_index(drop=True)
```

```python
scores = pd.concat([math, writing, reading], axis=1)
scores.columns = ['Math', 'Writing', 'Reading']
scores
```

```python
scores.plot(kind='scatter',x='Math', y='Writing')
```

```python
scores.plot(kind='scatter',x='Reading', y='Writing')
```

```python
math.corr(writing)
```

```python
reading.corr(writing)
```
